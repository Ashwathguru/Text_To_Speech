```python
def text_to_speech(textfile,audiofile):
    from gtts import gTTS
    import os
    with open(textfile, 'r') as file:
        data = file.read().replace('\n', '')
    #print(data)
    language = 'en'
    audio = gTTS(text=data, lang=language, slow=False)
    audio.save(audiofile)
```

```python
def play_audio(audiofile):
    from playsound import playsound
    playsound(audiofile)
```

```python
textfile="D:\\text_to_speech\\far.txt"
audiofile="far.mp3"
text_to_speech(textfile,audiofile)
```

```python
play_audio(audiofile)
```
